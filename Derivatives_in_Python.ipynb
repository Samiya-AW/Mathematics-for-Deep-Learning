## Numerical Derviative
Basic Intuition

f ' (x) = (f(x + h) - f(x)) / h


```python
# Define derivative function

def derivative(f, x, h = 1e-5):
  return (f(x + h) - f(x))/h

# Define function of which derivative needs to be calculated

def f(x):
  return x**2

derivative(f, 2)
```

The derivative calculated above is not exactly equals to 4 due to 'h' since computers can't take a literal limit 0, it is approximated by using a very small 'h' instead of an infinitesimally small one.

To overcome this problem we import a library called `'SymPy'` used for symbolic mathematical calculations in Python.


## Exact Derivative

```python
# import SymPy library
import sympy as sp
x = sp.symbols('x')
f = 3*x**2
df = sp.diff(f, x)
print(df)
func = (3*x**3 + 7*x) - (6*x**4)
print(sp.diff(func, x))

# define a function that takes a function from user to differentiate

def func():
  f = input("Write the function to differentiate: ")

  # return the derivative of the function given by user
  return sp.diff(f, x)

func()
```

## Partial Derivative

```python
x, y = sp.symbols('x y')
f = x**4 + y**3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
print(f"Derivative of f = x**4 + y**3 w.r.t x = {df_dx}\nDerivative of f = x**4 + y**3 w.r.t y = {df_dy}")

# Partial derivative of another function
f = 12*y + 34*x
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
print(f"\nDerivative of f = 3*x + 5*y w.r.t x = {df_dx}\nDerivative of f = 3*x + 5*y w.r.t y = {df_dy}")
```

## Chain Rule

```python
# Find derivative of 'y' w.r.t to 'x'

x, y, u = sp.symbols('x y u')
y = u**4
dy_du = sp.diff(y, u)
print(f"Derivative of y = u**4 w.r.t 'u' = {dy_du}")
u = x + 1
du_dx = sp.diff(u, x)
print(f"Derivative of u = x + 1 w.r.t 'x' = {dy_du}")

# Chain Rule = dy_dx = dy_du * du_dx
# Apply chain rule

dy_dx = dy_du * du_dx
print('\nAfter applying chain rule:')
print(f"Derivative of y = u**4 w.r.t 'x' = {dy_dx}")
```

```python
# Change the functions of y and u and then find derivative of y w.r.t 'x'
x, y, u = sp.symbols('x y u')
y = 4*u**3
dy_du = sp.diff(y, u)
print(f"Derivative of y = 4*u**3 w.r.t 'u' = {dy_du}")
u_exp = x**2
du_dx = sp.diff(u_exp, x)
print(f"Derivative of u = x**2 w.r.t 'x' = {dy_du}")

# Apply chain rule

dy_dx = dy_du * du_dx
print('\nAfter applying chain rule:')
print(f"Derivative of y = 4*u**3 w.r.t 'x' = {dy_dx}")
```

```python
# Further simplify the derivative by substituting the value of u by its experession 'x'
dy_dx = dy_du.subs(u, u_exp) * du_dx
print(dy_dx)
```

## Gradient
Gradient is a vector that contains all partial derivaties of a function written as follows in mathematical terms:

$\nabla f(x, y) = <4x^3, 3y^2>$ (from above partial derivatives cell)

```python
x, y = sp.symbols('x y')
f = x**2 + y**2

# Create a vector name 'grad' to store all the partial derivatives of f
grad = [sp.diff(f, var) for var in (x, y)]
print(f"Gradient = {grad}")
```

```python
# New function with 3 inputs

x, y, z = sp.symbols('x y z')
f = 6*x**2 + 4*y**2 + z

# Create vector named 'grad' to store partial derivatives of f
grad = []
for var in (x, y, z):
  grad.append(sp.diff(f, var))
  print(f"Gradient = {grad}")
```
