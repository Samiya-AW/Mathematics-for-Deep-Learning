**What is NumPy?**

NumPy = Numerical Python

NumPy is a Python library used to work with vectors, matrices and tensors efficiently.

###**Vector**

```python
# import NumPy library

import numpy as np
```

```python
# Create two vectors a and b

a = np.array([1, 2, 3])
print(a)
b = np.array([3, 4, 5])
print(b)
```

```python
# Addition of two vectors a + b
a + b
```

```python
# Subtraction of two vectors a - b
a - b
```

```python
# Increase the length of vector a by 2
# Do so by multiplying a by scalar value 2
# Scalar product of a vector

a * 2
```

```python
# Dot product of two vectors a and b

# Basic intuition
(a[0] * b[0]) + (a[1] * b[1]) + (a[2] * b[2])

# Using NumPy
dot_product = np.dot(a, b)
print(dot_product)
```

```python
# Magnitude and unit of a vector

# Basic intuition of magnitude
m_a = ((a[0]**2) + (a[1]**2) + (a[2]**2))**0.5
print(m_a)

# Using NumPy
m = np.linalg.norm(a)
print(m)

# Unit Vector
uv_a = a / m
print(uv_a)

# Magnitude of resultant unit vector (uv_a) will be equal to 1
print(np.linalg.norm(uv_a))
```

###**Matrices**

```python
# Create two matrices named A and B

A = np.array([[1, 2], [3, 4]])
B = np. array([[1, 0], [5, 6]])
A.shape, B.shape
```

```python
# Addition of two matrices A and B
A + B
```

```python
# Subtraction of two matrices A and B
A - B
```

```python
# Multiplication of two matrices A and B
print(A.shape, B.shape)
A @ B

# OR

np.dot(A, B)
```

```python
# Scalar multiplication of matrix A
3 * A
```

```python
# Determinant of matrix A

# Basic intuition
det = ((A[0][0] * A[1][1]) - (A[0][1] * A[1][0]))
print(det)

# Using NumPy
det_A = np.linalg.det(A)
print(det_A)
```
